```python
x = pd.merge(x_order_report, x_sku, on = 'SKU')
x.head()
```

```python
x['Weight (g)'].unique()
```

```python
x.rename(columns = {'Weight (g)':'weight_x'}, inplace = True)
x.rename(columns = {'ExternOrderNo':'Order ID'}, inplace = True)
x['weight_x'] = x['weight_x']/1000
def slab_x(n):
    if n<=0.5:
        return 0.5
    elif n>0.5 and n<=1:
        return 1
x['weight_slab_x']= x['weight_x'].map(slab_x)
x.head()
```

```python
x['weight_slab_x'].max()
```

```python
company= pd.merge(x,courier_company, on='Order ID')
company.head()
```

```python
x_pincode.rename(columns={'Zone':'zone_x'}, inplace = True)
company.rename(columns={'Zone':'zone_c'}, inplace = True)
final = pd.merge(x_pincode,company, on = 'Warehouse Pincode')
final['weight_c']= final['weight_x']
final.head()
```

```python
final.shape
```

```python
count=0
for i,j in zip(final['zone_x'], final['zone_c']):
    if i!=j:
        count+=1       
print(count)
```

```python
def slab_charged(a):
    if a<=0.5:
        return 0.5
    elif a>0.5 and a<=1:
        return 1
    elif 1<a<=1.5:
        return 1.5
    elif 1.5<a<=2:
        return 2
    elif 2<a<=2.5:
        return 2.5
    elif 2.5<a<=3:
        return 3
    elif 3<a<=3.5:
        return 3.5
    elif 3.5<a<=4:
        return 4
    elif 4<a<=4.5:
        return 4.5
final["weight_slab_c"]=final['Charged Weight'].map(slab_charged)
```

```python
final["weight_slab_x"] = final["weight_slab_x"]*final["Order Qty"]
```

```python
x = final["weight_slab_x"]
y = final["zone_x"]
z = final["Type of Shipment"]
def func(a,b,c):
    new = []
    for i in zip(a,b,c):
        ## for zone a
        if i[1]=='a' and i[-1]== 'Forward charges' and i[0]<=0.5:
            new.append(29.5)
        elif i[1]=='a' and i[-1]== 'Forward charges' and i[0]>0.5:
            new.append(29.5 + (i[0]/0.5-1)*23.6)
        elif i[1]=='a' and i[-1]== 'Forward and RTO charges' and i[0]<=0.5:
            new.append(29.5+13.6)
        elif i[1]=='a' and i[-1]== 'Forward and RTO charges' and i[0]>0.5:
            new.append(29.5+13.6+ (i[0]/0.5-1)*23.6 + (i[0]/0.5-1)*23.6)
        ## for zone b
        elif i[1]=='b' and i[-1]== 'Forward charges' and i[0]<=0.5:
            new.append(33)
        elif i[1]=='b' and i[-1]== 'Forward charges' and i[0]>0.5:
            new.append(33 + (i[0]/0.5-1)*28.3)
        elif i[1]=='b' and i[-1]== 'Forward and RTO charges' and i[0]<=0.5:
            new.append(33+20.5)
        elif i[1]=='b' and i[-1]== 'Forward and RTO charges' and i[0]>0.5:
            new.append(33+20.5+ (i[0]/0.5-1)*28.3 + (i[0]/0.5-1)*28.3)
        ## for zone c
        elif i[1]=='c' and i[-1]== 'Forward charges' and i[0]<=0.5:
            new.append(40.1)
        elif i[1]=='c' and i[-1]== 'Forward charges' and i[0]>0.5:
            new.append(40.1 + (i[0]/0.5-1)*38.9)
        elif i[1]=='c' and i[-1]== 'Forward and RTO charges' and i[0]<=0.5:
            new.append(40+31.9)
        elif i[1]=='c' and i[-1]== 'Forward and RTO charges' and i[0]>0.5:
            new.append(40+31.9+ (i[0]/0.5-1)*38.9 + (i[0]/0.5-1)*38.9)
        ## for zone d
        elif i[1]=='d' and i[-1]== 'Forward charges' and i[0]<=0.5:
            new.append(45.4)
        elif i[1]=='d' and i[-1]== 'Forward charges' and i[0]>0.5:
            new.append(45.4 + (i[0]/0.5-1)*44.8)
        elif i[1]=='d' and i[-1]== 'Forward and RTO charges' and i[0]<=0.5:
            new.append(45.4+41.3)
        elif i[1]=='d' and i[-1]== 'Forward and RTO charges' and i[0]>0.5:
            new.append(45.4+41.3+ (i[0]/0.5-1)*44.8 + (i[0]/0.5-1)*44.8)
        ## for zone e
        elif i[1]=='e' and i[-1]== 'Forward charges' and i[0]<=0.5:
            new.append(56.6)
        elif i[1]=='e' and i[-1]== 'Forward charges' and i[0]>0.5:
            new.append(56.6 + (i[0]/0.5-1)*55.5)
        elif i[1]=='e' and i[-1]== 'Forward and RTO charges' and i[0]<=0.5:
            new.append(56.6+50.7)
        elif i[1]=='e' and i[-1]== 'Forward and RTO charges' and i[0]>0.5:
            new.append(56.6+50.7+ (i[0]/0.5-1)*55.5 + (i[0]/0.5-1)*55.5)
    return new
final["expected charges as per x"] = func(x,y,z)
final.head()
```
